## UFO Sightings DataSet

The following data was taken from the National UFO Reporting Center http://www.nuforc.org/ and represents 5 months of reported UFO sightings from January 2016 to May 2016. 

Let's Perform a data analysis on 3 months of data by answering the questions below


```python
import pandas as pd

jan_ufo = pd.read_csv('https://raw.githubusercontent.com/mafudge/datasets/master/ufo-sightings/ufo-sightings-2016-01.csv')
jan_ufo.head()
```

```python
# in this cell write code toread in the sightings from feb (2016-01) and march (2016-02) into variables
feb_ufo = pd.read_csv('https://raw.githubusercontent.com/mafudge/datasets/master/ufo-sightings/ufo-sightings-2016-02.csv')
mar_ufo = pd.read_csv('https://raw.githubusercontent.com/mafudge/datasets/master/ufo-sightings/ufo-sightings-2016-03.csv')
```

```python
# next combine all three dataframes into a single dataframe, UFO.  Here's the docs:
# http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.append.html
# if you ufo.count() you should see 1229 rows.
ufo = jan_ufo.append(feb_ufo).append(mar_ufo)
ufo.count()
```

Let's answer some questions!

```python
# get a summary of sightings by state
ufo['State'].value_counts()
```

```python
# are there any Allentown, PA sightings?  List them.
ufo [ (ufo['City'] =='Allentown') & (ufo['State']=='PA') ]
```

```python
#  list sightings have "orange" in the summary. We use .str.find to locate the word orange" somewhere in the summary
orange = ufo[ ufo['Summary'].str.find('orange')>=0 ]
orange
```

```python
# finally write a program to ask for a search term such as "orange" or "formation" and then returns the sightings
# with that term in the summary
print("Welecome to UFO World... You are not alone!")
search = input("Enter Search Term: ")
results = ufo[ ufo['Summary'].str.find(search)>=0 ]
results
```
